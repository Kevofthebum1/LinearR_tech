# Linear Regression

```python
#We will drop the columns to build our linear model

#independent variables
X = cancer.drop(columns={'Prostate_C', 'Urinary_Sy'})

#dependent variable
y = cancer['Prostate_C']
```

```python
#Then we want to split the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=1)
```

```python
#Creating a linear regression model using statsmodels OLS and interpretate coeffiecients

#importing statsmodel
from statsmodels.graphics.gofplots import ProbPlot
from statsmodels.formula.api import ols
import statsmodels.api as sm

#Add the inercept to the data
X_train_ols = sm.add_constant(X_train)
X_test_ols = sm.add_constant(X_test)

#create the model

#defining the ols model
model1 = sm.OLS(y_train, X_train_ols).fit()

#summary of model1
model1.summary()
```

* We need to check for the `p-value` to check which are statistically significant or have predictive power to predict the target variable since not all the variables that were mentioned above are statistically significant to predict the outcome variable.

- **Interpreting the Regression Results:**

1. **Adjusted. R-squared**: It reflects the fit of the model.
    - R-squared values range from 0 to 1, where a higher value generally indicates a better fit, assuming certain conditions are met.
    
2. **coeff**: It represents the change in the output Y due to a change of one unit in the variable (everything else held constant).
3. **std err**: It reflects the level of accuracy of the coefficients.
    - The lower it is, the more accurate the coefficients are.
4. **P >|t|**: It is p-value.
   
   * Pr(>|t|) : For each independent feature there is a null hypothesis and alternate hypothesis 

    Ho : Independent feature is not significant 
   
    Ha : Independent feature is significant 
    
   * A p-value of less than 0.05 is considered to be statistically significant.

   
5. **Confidence Interval**: It represents the range in which our coefficients are likely to fall (with a likelihood of 95%).

* To be able to make statistical inferences from our model, **we will have to test the significance of the regression coefficients and linear regression assumptions.**

```python
#Checking for the performance of the train and test data set

#RSME
def rmse(predictions, targets):
    return np.sqrt((targets-predictions)**2).mean()

#MAPE
def mape(predictions, targets):
    return np.mean(np.abs((targets-predictions))/targets)*100

#MAE
def mae(predictions, targets):
    return np.mean(np.abs((targets-predictions)))

#Model performance of the train and test data
def model_performance(olsmodel, X_train, X_test, y_train, y_test):
    
    #In-sample predictions (train data)
    y_pred_train = olsmodel.predict(X_train)
    y_observed_train = y_train
    
    #Prediction on test data
    y_pred_test = olsmodel.predict(X_test)
    y_observed_test = y_test
    
    #Printing out the performance
    print(
        pd.DataFrame(
            {
                "Data": ["Train", "Test"],
                "RMSE": [
                    rmse(y_pred_train, y_observed_train),
                    rmse(y_pred_test, y_observed_test),
                ],
                "MAE": [
                    mae(y_pred_train, y_observed_train),
                    mae(y_pred_test, y_observed_test),
                ],
                "MAPE": [
                    mape(y_pred_train, y_observed_train),
                    mape(y_pred_test, y_observed_test),
                ],
            }
        )
    )
    
#Printing out the model performance
model_performance(model1, X_train_ols, X_test_ols, y_train, y_test)
```

* Based on our RMSE, MAE, and MAPE, We see that the model is **underfitting** since the test dataset are a bit higher than the train dataset.
* We would have to perform cross validation to see what is our RSquared

# Performing cross validation and comparing its average  performance to OLS performance

```python
#Importing the functions
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

#Building a regression model using Sklearn Linear regression
lr = LinearRegression()

#Need to a cross val score for cross validation
cv_Score11 = cross_val_score(lr, X_train_ols, y_train, cv=3) #cv represents data is divided into n folds
cv_Score12 = cross_val_score(lr, X_train_ols, y_train, cv=3, scoring='neg_mean_squared_error')

#Print the R-squared and Mean Squared Error
print("RSquared: %0.3f (+/- %0.3f)" % (cv_Score11.mean(), cv_Score11.std() * 2))
print("Mean Squared Error: %0.3f (+/- %0.3f)" % (-1*cv_Score12.mean(), cv_Score12.std() * 2))
```

* For our cross validation, we have that our **cv=3**. This would get us an R^2 to be in the range 0 and 1.

# Building a Decision Tree

```python
#We will try doing a Decision Tree Regressor using sklearn

from sklearn.tree import DecisionTreeRegressor
```

```python
#splitting the data in 70:30 ratio of train to test data
#seperate the independent and dependent variable
Y1 = cancer['Prostate_C']
X1 = cancer.drop(columns={'Prostate_C', 'Urinary_Sy'})

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, Y1, test_size=0.3, random_state=1)
```

```python
#Defining the Decision tree regressor
dt = DecisionTreeRegressor(random_state=1)

#Fitting Decison Tree regressor to train dataset
dt.fit(X_train1, y_train1)
```

```python
model_performance(dt, X_train1, X_test1, y_train1, y_test1)
```

* **The model seems to overfit the data** as rmse, mae, and mape value of the train data is 0, but the values of the test data is higher.

```python
#We can do a tree plot for the features

#import the plot tree
from sklearn.tree import plot_tree
```

```python
features = list(X1.columns)

plt.figure(figsize=(35,25))
plot_tree(dt, max_depth=4, feature_names=features,filled=True,fontsize=12,node_ids=True,class_names=True)
plt.show()
```

**We can analyze the variables after ploting the feature importance for each variable.**

```python
importances = dt.feature_importances_

columns = X1.columns
importance_df=pd.DataFrame(importances,index=columns,columns=['Importance']).sort_values(by='Importance',ascending=False)
plt.figure(figsize=(8,4))
sns.barplot(x=importance_df.Importance,y=importance_df.index)
```

# Building Random Forest

```python
#importing random forest regressor using sklearn

from sklearn.ensemble import RandomForestRegressor
```

#### Parameters for regression
**n_estimators**: The number of trees in the forest.

**min_samples_split**: The minimum number of samples required to split an internal node:

**max_depth**
The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

**max_features{“auto”, “sqrt”, “log2”, 'None'}**: The number of features to consider when looking for the best split.

- If “auto”, then max_features=sqrt(n_features).

- If “sqrt”, then max_features=sqrt(n_features) (same as “auto”).

- If “log2”, then max_features=log2(n_features).

- If None, then max_features=n_features.

```python
#defing the Random Forest regressor
rf = RandomForestRegressor(n_estimators=100, random_state=1)

#fitting the model
rf.fit(X_test1, y_test1)
```

```python
#checking model performance of the test dataset
rf.score(X_test1, y_test1)
```

```python
model_performance(rf, X_train1, X_test1, y_train1, y_test1)
```

# Checking the feature importance of each variable in Random Forest and comparing to Decision Tree

```python
importances = rf.feature_importances_

columns=X1.columns
importance_df=pd.DataFrame(importances,index=columns,columns=['Importance']).sort_values(by='Importance',ascending=False)
plt.figure(figsize=(8,4))
sns.barplot(x=importance_df.Importance,y=importance_df.index)
```
